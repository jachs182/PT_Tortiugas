```python
# --- Preparación ---
date_cols = [
'fecha_recolecta',
    'fecha_prob_eclosion',
 'fecha_emergencia',
    'hora_avistamiento',
    'fecha_hora_recolecta',
]

for col in date_cols:
    df_joined[col] = pd.to_datetime(df_joined[col], errors="coerce")
```

```python
#converir fechas a numeros

date_cols = [
"fecha_recolecta"
]

for col in date_cols:
    df_joined[col] = df_joined[col].astype('int64') / 10**9
```

```python
df_joined.info()
```

```python
def gdm_to_dd(coord_str):
    import re
    # Extraer números usando regex
    parts = re.findall(r"[-+]?\d*\.\d+|\d+", coord_str)
    if len(parts) < 4: return None, None
    
    # Latitud: Grados + (Minutos/60)
    lat = float(parts[0]) + float(parts[1])/60
    # Longitud: (Grados + (Minutos/60)) * -1 si es Oeste (W)
    lon = (float(parts[2]) + float(parts[3])/60) * -1
    return lat, lon
```

```python
df_joined[['lat', 'lon']] = df_joined['GPS'].apply(lambda x: pd.Series(gdm_to_dd(x)))
```

```python
# Convertir todo a mayúscula inicial para evitar duplicados por formato
df_joined['especie'] = df_joined['especie'].str.capitalize()
```

```python
# Convertir la columna ESPECIE en columnas numéricas (Dummies)
df_procesado = pd.get_dummies(df_joined, columns=['especie'], prefix='ESP')
```

```python
cols_to_drop = [
    'Unnamed: 0',
    'id',
    'hora_avistamiento',
    'se_observo',
 'lscc_cm',
 'acc_cm',
 'hora_recolecta',
 'tamano_nidada',
 'hora_siembra',
 'huevos_sembrados',
 'fecha_prob_eclosion',
 'fecha_emergencia',
 'crias_vivas',
 'crias_muertas',
 'hsda',
 'hcda',
 'total_crias',
 'crias_deformes',
 'dias_incubacion',
    'aleta_izq_marca_palace',
 'aleta_der_marca_palace',
 'no_serie_marca_palace',
 'recaptura_palace',
 'no_serie_recaptura',
 'aleta_izq_marca_externa',
 'aleta_der_marca_externa',
 'procedencia_marca_externa',
 'secuencia_fotografica_camara',
 'observaciones',
 'fecha_hora_recolecta',
    'time',
 'tp_mm',
 'ssrd_wm2',
 'swh',
 'mwd',
 'mwp',
 'shts',
 'tp',
 'ssrd',

]

df_ml = df_procesado.drop(columns=cols_to_drop)
```

## Visualizacion de los nidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear el DataFrame limpio (sin los ceros)
df_cleaned = df_ml[(df_ml['lat'] != 0) & (df_ml['lon'] != 0)].copy()

# 2. Asegurar que las coordenadas sean numéricas
df_cleaned['lat'] = df_cleaned['lat'].astype(float)
df_cleaned['lon'] = df_cleaned['lon'].astype(float)

# 3. Graficar
plt.figure(figsize=(12, 8))

# Usamos 'fecha_recolecta' o 'estacion' para el color para ver el desplazamiento temporal
scatter = sns.scatterplot(
    data=df_cleaned, 
    x='lon', 
    y='lat', 
    hue='estacion',  # O usa 'estacion' si prefieres ver por temporadas
    palette='viridis', 
    alpha=0.7, 
    s=60
)

plt.title('Distribución Geográfica de Nidos (Datos Limpios)', fontsize=14)
plt.xlabel('Longitud', fontsize=12)
plt.ylabel('Latitud', fontsize=12)
plt.legend(title='Fecha / Periodo', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

## DBSCAN Geoespacial

```python
import numpy as np
from sklearn.cluster import DBSCAN

# 1. Convertir a radianes para Haversine
coords = np.radians(df_clean[['lat', 'lon']].values)

# 2. Configurar DBSCAN
# eps: radio máximo. 0.1 / 6371 approx = 100 metros
kms_per_radian = 6371.0088
epsilon = 0.1 / kms_per_radian 

db = DBSCAN(eps=epsilon, min_samples=5, algorithm='ball_tree', metric='haversine')
df_clean['cluster_id'] = db.fit_predict(coords)
```

```python
df_clean['cluster_id'].head()
```

```python
print(f"Nidos analizados: {len(df_clean)}")
print(f"Clusters detectados: {len(set(df_clean['cluster_id'])) - (1 if -1 in df_clean['cluster_id'].values else 0)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.title('Identificación de Zonas de Anidación')
# Graficar nidos normales
sns.scatterplot(data=df_clean[df_clean['cluster_id'] != -1], x='lon', y='lat', 
                hue='cluster_id', palette='tab10', s=100, alpha=0.6)
```

```python
# Graficar clusters y nidos aislados juntos para ver la diferencia
plt.figure(figsize=(10, 6))

# Dibujar los nidos estables (clusters) en gris claro para que no distraigan
sns.scatterplot(data=df_clean[df_clean['cluster_id'] != -1], 
                x='lon', y='lat', color='blue', alpha=0.5, label='Zonas Estables')

# Dibujar los nidos aislados (los que te interesan) en rojo fuerte
sns.scatterplot(data=df_clean[df_clean['cluster_id'] == -1], 
                x='lon', y='lat', color='red', marker='X', s=100, label='Desplazamiento/Atípicos')

plt.title('Nidos Aislados vs Zonas de Anidación Tradicionales')
plt.show()
```

## Analisis por especie

## Tortuga Blanca

```python
# Ejemplo para analizar solo la tortuga Blanca (Cm)
df_cm = df_cleaned[df_cleaned['ESP_Cm'] == True].copy()
coords_cm = np.radians(df_cm[['lat', 'lon']].values)

# Correr DBSCAN solo para Cm
db_cm = DBSCAN(eps=epsilon, min_samples=5, metric='haversine')
df_cm['cluster_id'] = db_cm.fit_predict(coords_cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear una columna temporal para comparar "Agrupado" vs "Aislado"
df_cm['tipo_nido'] = df_cm['cluster_id'].apply(lambda x: 'Aislado' if x == -1 else 'Agrupado')

# Comparar Temperatura del Suelo (stl1) entre tipos de nido
plt.figure(figsize=(10, 6))
sns.boxplot(x='tipo_nido', y='stl1_celsius', data=df_cm)
plt.title('Influencia de la Temp. del Suelo en la Dispersión (Tortuga Blanca)')
plt.show()
```

```python
# Correlación de variables ambientales con el hecho de ser un nido "Aislado"
# Creamos una variable booleana: 1 si es aislado, 0 si está en cluster
df_cm['es_aislado'] = (df_cm['cluster_id'] == -1).astype(int)

cols_interes = ['stl1_celsius', 't2m_celsius', 'wind_speed_10m_ms', 'es_aislado']
print(df_cm[cols_interes].corr()['es_aislado'].sort_values(ascending=False))
```

## Tortuga Caguama

```python
# Ejemplo para analizar solo la tortuga Caguama (Cc)
df_cc = df_cleaned[df_cleaned['ESP_Cc'] == True].copy()
coords_cc = np.radians(df_cc[['lat', 'lon']].values)

# Correr DBSCAN solo para Cc
db_cc = DBSCAN(eps=epsilon, min_samples=5, metric='haversine')
df_cc['cluster_id'] = db_cc.fit_predict(coords_cc)
```

```python
# Crear una columna temporal para comparar "Agrupado" vs "Aislado"
df_cc['tipo_nido'] = df_cc['cluster_id'].apply(lambda x: 'Aislado' if x == -1 else 'Agrupado')

df_cc['tipo_nido'].unique()

```

```python
# Comparar Temperatura del Suelo (stl1) entre tipos de nido
plt.figure(figsize=(10, 6))
sns.boxplot(x='tipo_nido', y='stl1_celsius', data=df_cc)
plt.title('Influencia de la Temp. del Suelo en la Dispersión (Tortuga Caguama)')
plt.show()
```

```python
# Correlación de variables ambientales con el hecho de ser un nido "Aislado"
# Creamos una variable booleana: 1 si es aislado, 0 si está en cluster
df_cc['es_aislado'] = (df_cc['cluster_id'] == -1).astype(int)

cols_interes = ['stl1_celsius', 't2m_celsius', 'wind_speed_10m_ms', 'es_aislado']
print(df_cc[cols_interes].corr()['es_aislado'].sort_values(ascending=False))
```

# Analisis por zona geografica

```python
# Unimos la información de las etiquetas de clusters de vuelta al DF principal
# Suponiendo que ya corriste el DBSCAN para cada uno:
df_clean['tipo_nido'] = 'Aislado' # Por defecto todos son aislados

# Actualizar con los que sí están agrupados según tus análisis previos
df_clean.loc[df_cm[df_cm['cluster_id'] != -1].index, 'tipo_nido'] = 'Agrupado'
df_clean.loc[df_cc[df_cc['cluster_id'] != -1].index, 'tipo_nido'] = 'Agrupado'
```

```python
# Crear columna de especie única basada en las booleanas
df_clean['especie_label'] = 'Otra'
df_clean.loc[df_clean['ESP_Cc'] == True, 'especie_label'] = 'Caguama'
df_clean.loc[df_clean['ESP_Cm'] == True, 'especie_label'] = 'Blanca'
```

```python
# Ver en qué zonas hay más nidos aislados
zona_desplazamiento = df_clean.groupby(['zona', 'ESP_Cc', 'ESP_Cm', 'tipo_nido']).size().unstack(fill_value=0)
zona_desplazamiento['%_aislados'] = (zona_desplazamiento['Aislado'] / (zona_desplazamiento['Aislado'] + zona_desplazamiento['Agrupado'])) * 100

print(zona_desplazamiento.sort_values('%_aislados', ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# Graficamos la evolución de la longitud a través de los años
sns.boxplot(data=df_clean, x='year', y='lon', hue='especie_label', palette='Set2')

plt.title('Evidencia de Desplazamiento: Cambio en la Longitud de los Nidos (2018-2024)')
plt.ylabel('Longitud (Valores menores = Más tierra adentro)')
plt.xlabel('Año de Anidación')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Comparar la latitud y longitud promedio por zona
analisis_zonas = df_clean.groupby(['zona', 'especie_label', 'tipo_nido'])[['lat', 'lon']].mean().reset_index()

# Visualizar la ubicación de los nidos "Aislados" vs "Agrupados" por zona
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_clean, x='lon', y='lat', hue='zona', style='tipo_nido', palette='deep', s=100)

plt.title('Distribución Espacial por Zona y Estabilidad del Nido')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
